```python
import pandas as pd
import json
```

```python
# the xlsx cab be obtained at 
# https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/87YG4K
# and then change the name to visits.xlsx and place it in the same folder as the notebook file
df = pd.read_excel("./visits.xlsx", index_col="TripID")
```

```python
# get visit counts per-year
df.groupby("TripYear").size().reset_index(name="visitCount").to_json('visits/visits-by-year.json', orient='records')
```

```python
# get countries that made most diplomatic travels in 2020 and also countries that received most diplomatic visits on the same year.
trips_in_2020 = df[df['TripYear'] == 2020]
top_visitors = trips_in_2020.groupby("LeaderCountryOrIGO").size().sort_values(ascending=False).reset_index(name="visitCount").rename(columns={"LeaderCountryOrIGO": "country"})
top_visited = trips_in_2020.groupby("CountryVisited").size().sort_values(ascending=False).reset_index(name="visitCount").rename(columns={"CountryVisited": "country"})

output = {
    "topVisitors": top_visitors.to_dict('records'),
    "topVisited": top_visited.to_dict('records'),
    "year": 2020
}

with open('visits/diplomatic-activity-2020.json', 'w') as f:
    json.dump(output, f)
```
