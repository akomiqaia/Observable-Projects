```python
import pandas as pd
import json
```

```python
# the xlsx cab be obtained at 
# https://dataverse.harvard.edu/dataset.xhtml?persistentId=doi:10.7910/DVN/87YG4K
# and then change the name to visits.xlsx and place it in the same folder as the notebook file
df = pd.read_excel("./visits.xlsx", index_col="TripID")
```

```python
df_new_cols = df.rename(columns={"LeaderCountryOrIGO": "CountryOrigin" })
df_new_cols.groupby(["TripYear", "CountryOrigin", "CountryVisited"]).agg(Count=("TripYear", "count")).reset_index().to_json("visits/visits-by-year.json", orient="records")
```
